```python
##Steps we need to train the algorithm
##1. split the data  training and testing data
##2.solve linear regression
##3.making the predictions  
```

```python
#Predict the values b/w attributes and labels i.e.X and y
#while attribute are independent variables and labels are dependent variables

X= data.iloc[:,:-1].values
y=data.iloc[:,:-1].values
```

```python
#next thing we need to import the model use to training and testing data
# this model is called as psychic module so the data must be splitted in some more parts
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0) ##this will 80% of training data 
                                                                                    #and 20% of testing data
```

```python
##next we need given data must be trained algorithms
## we need to call linear regression class
##next, we need to fit the model along with the training data

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
#It is easy is implement for cyclic module linear regression algorithm
#it is the best way to find the best way of find the solution to intercept and slope
##intercept --> it results the line that fits the data to see the value find the linear regression with the help of these cmd

print(regressor.intercept_)
```

```python
print(regressor.coef_)
```

```python
y_pred =regressor.predict(X_test)
```

```python
print(y_pred)
```

```python
df =pd.DataFrame({'Actual':[y_test],'Predict':[y_pred]})
print(df)
```
